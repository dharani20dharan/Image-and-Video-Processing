# Experiment 1: SIFT Feature Matching

## Problem
To detect keypoints using SIFT and display best matches between two images.

## Theory
SIFT (Scale-Invariant Feature Transform) detects keypoints invariant to scale and rotation.
Matching is done using descriptors.

```python
import cv2
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread('image1.jpeg', 0)
img2 = cv2.imread('image2.jpeg', 0)

# Create SIFT object
sift = cv2.SIFT_create()

# Detect keypoints and descriptors
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)

# Matcher
bf = cv2.BFMatcher()

matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)

# Draw top 10 matches
matched_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], None, flags=2)

plt.figure(figsize=(12,6))
plt.imshow(matched_img, cmap='gray')
plt.title("Top 10 Matches")
plt.axis('off')
plt.show()
```

# Experiment 2: 3D Reconstruction using COLMAP

## Problem
To reconstruct a 3D scene from multiple images.

## Theory
COLMAP uses Structure-from-Motion (SfM) to estimate camera poses and reconstruct 3D points.

### Steps for 3D Reconstruction:

1. Install COLMAP software
2. Load multiple images of the same scene
3. Extract features
4. Match features across images
5. Perform Structure-from-Motion
6. Generate sparse 3D model
7. Perform dense reconstruction
8. Visualize 3D output
