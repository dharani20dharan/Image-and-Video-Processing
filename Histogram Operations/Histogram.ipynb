# Experiment 1: Histogram of an Image

## Problem
To compute and plot the histogram of a grayscale image.

## Theory
A histogram represents the distribution of pixel intensities in an image.
X-axis → Intensity values (0–255)  
Y-axis → Frequency of pixels

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.png', 0)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(img.ravel(), bins=256, range=[0,256])
plt.title("Histogram")

plt.show()
```

# Experiment 2: Histogram Equalization

## Problem
To enhance image contrast using histogram equalization.

## Theory
Histogram equalization redistributes pixel intensities to improve contrast, especially in low-contrast images.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.png', 0)

equalized = cv2.equalizeHist(img)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title("Equalized")
plt.axis('off')

plt.show()
```

# Experiment 3: Histogram Modification (Matching)

## Problem
To modify the histogram of one image to match another image.

## Theory
Histogram matching adjusts the pixel distribution of one image to resemble another.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms

img1 = cv2.imread('image.png', 0)
img2 = cv2.imread('image2.jpg', 0)

matched = match_histograms(img1, img2)

plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(img1, cmap='gray')
plt.title("Source")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img2, cmap='gray')
plt.title("Reference")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched, cmap='gray')
plt.title("Matched")
plt.axis('off')

plt.show()
```

# Experiment 4: Image Transformation

## Problem
To perform image reflection.

## Theory
Reflection flips the image along horizontal or vertical axes.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

horizontal = cv2.flip(img_rgb, 1)
vertical = cv2.flip(img_rgb, 0)

plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(horizontal)
plt.title("Horizontal Flip")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(vertical)
plt.title("Vertical Flip")
plt.axis('off')

plt.show()
```

# Experiment 5: 3-bit Image Generation

## Problem
To generate a low-bit image and modify histogram.

## Theory
Reducing bit levels reduces intensity variations. A 3-bit image has only 8 intensity levels.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.png', 0)

# 3-bit image (8 levels)
img_3bit = (img // 32) * 32

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_3bit, cmap='gray')
plt.title("3-bit Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(img_3bit.ravel(), bins=8)
plt.title("Histogram")

plt.show()
```

# Experiment 6: Image Quality Metrics

## Problem
To compute PSNR, SSIM, and LPIPS between two images.

## Theory
PSNR → Measures reconstruction quality  
SSIM → Measures structural similarity  
LPIPS → Measures perceptual similarity

```python
import cv2
from skimage.metrics import peak_signal_noise_ratio, structural_similarity

img1 = cv2.imread('image.png', 0)
img2 = cv2.GaussianBlur(img1, (5,5), 0)

psnr = peak_signal_noise_ratio(img1, img2)
ssim = structural_similarity(img1, img2)

print("PSNR:", psnr)
print("SSIM:", ssim)
```
