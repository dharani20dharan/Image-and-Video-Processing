# Lab 4: Spatial Filtering and Statistical Foundations

## Theory

### Convolution
Convolution involves flipping the kernel and performing a weighted sum over the neighborhood.

g(x, y) = ΣΣ w(s,t) f(x−s, y−t)

---

### Correlation
Correlation applies the kernel directly without flipping.

g(x, y) = ΣΣ w(s,t) f(x+s, y+t)

---

### Difference
- Convolution → kernel flipped
- Correlation → no flipping

For symmetric kernels, both produce identical results.

---

### Central Limit Theorem (CLT)
The CLT states that the mean of a large number of independent random variables tends toward a Gaussian distribution.

This explains:
- Why Gaussian noise appears naturally
- Why repeated averaging produces smooth images

# Experiment 1: Spatial Filtering using Convolution

## Problem
To apply convolution on an image using a filter kernel.

## Theory
Convolution involves sliding a kernel over the image and computing weighted sums.
It is used for blurring, sharpening, and edge detection.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg', 0)

# Example kernel (blur)
kernel = np.ones((3,3), np.float32) / 9

conv = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(conv, cmap='gray')
plt.title("Convolution (Blur)")
plt.axis('off')

plt.show()
```

# Experiment 2: Spatial Filtering using Correlation

## Problem
To apply correlation on an image.

## Theory
Correlation is similar to convolution but does not flip the kernel.
It is used for template matching and feature detection.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg', 0)

kernel = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])

# Correlation (same as filter2D in OpenCV)
corr = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(corr, cmap='gray')
plt.title("Correlation Output")
plt.axis('off')

plt.show()
```

# Experiment 3: Verification of Central Limit Theorem

## Problem
To verify CLT using random variables.

## Theory
Central Limit Theorem states that the sum of independent random variables tends toward a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate uniform random numbers
data = np.random.uniform(0, 1, (1000, 10))

# Take mean across axis
means = np.mean(data, axis=1)

plt.figure(figsize=(8,5))
plt.hist(means, bins=30)
plt.title("CLT Verification (Mean Distribution)")
plt.show()
```

Convolution and correlation are spatial filtering techniques used for modifying images. Convolution involves kernel flipping, while correlation does not. The Central Limit Theorem is verified as the distribution of sample means approaches a normal distribution.
